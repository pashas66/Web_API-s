```python
#Now we will pass `retrieve_NYT_json_data` as a single attribute to`urllib2.urlopen()` in the newly created `data` variable. and load the json data in to the `NYT_top_stories` variable.
data = urllib2.urlopen(retrieve_NYT_json_data).read()
NYT_top_stories = json.loads(data)
NYT_top_stories #JSON results
```
